# Classification des données Iris

```python
from sklearn.datasets import load_iris
data = load_iris()
```

```python
X, y = data.data, data.target
```

```python
import scipy
import numpy as np
scipy.stats.describe(data.data[:,:5])
```

```python
np.bincount(data.target)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
plt.figure(figsize=(20,10))
plot_tree(model, filled=True)
```

```python
model1 = DecisionTreeClassifier(max_depth = 3)
model1.fit(X_train, y_train)
```

```python
plot_tree(model1, filled=True)
```

```python
model2 = DecisionTreeClassifier(min_samples_leaf = 20)
model2.fit(X_train, y_train)
```

```python
plot_tree(model2, filled=True)
```

Le problème ici étant particulièrement simple, refaites une division apprentissage/test avec 5% des données en apprentissage et 95% test.

Calculez le taux d’éléments mal classifiés sur l’ensemble de test.

Faites varier (ou mieux, réalisez une recherche par grille avec GridSearchCV) les valeurs des paramètres max_depth et min_samples_leaf pour mesurer leur impact sur ce score.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
for mdepth in [1,2,3,4,5,6,7]:
    clf = DecisionTreeClassifier(max_depth=mdepth)
    clf.fit(X_train, y_train)
    print(clf.score(X_test, y_test))

for msplith in [2,3,5,10,15,20]:
    clf = DecisionTreeClassifier(min_samples_split=msplith)
    clf.fit(X_train, y_train)
    print(clf.score(X_test, y_test))
```

Pour chercher de bonnes valeurs pour les paramètres on peut utiliser la validation croisée avec GridSearchCV :

```python
from sklearn.model_selection import GridSearchCV

param = {"max_depth":[1,2,3,4,5,6,7],
"min_samples_split":[2,3,5,10,15,20]}
grid = GridSearchCV(model, param, cv=10)
grid.fit(X_train, y_train)
grid.best_score_
```

```python
grid.best_params_
```

```python
grid.best_estimator_.score(X_test, y_test)
```

# Affichage de la surface de décision

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_classes = 3
plot_colors = "bry"
plot_step = 0.02
```

```python
pair = [2,3]
```

```python
X = data.data[:, pair]
y = data.target
```

```python
clf = DecisionTreeClassifier().fit(X, y)
```

```python
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step), np.arange(y_min, y_max, plot_step))
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
cs = plt.contourf(xx, yy, Z, cmap=plt.cm.Paired)
plt.xlabel(data.feature_names[pair[0]])
plt.ylabel(data.feature_names[pair[1]])
plt.axis("tight")
```

```python
for i, color in zip(range(n_classes), plot_colors):
    idx = np.where(y == i)
    plt.scatter(X[idx, 0], X[idx, 1], c=color, label=data.target_names[i], cmap=plt.cm.Paired)
plt.axis("tight")
plt.suptitle("Decision surface of a decision tree using paired features")
plt.legend()
plt.show()
```
